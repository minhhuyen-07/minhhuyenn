```python
#W6A10:
lst = list(map(int,input().split()))
k = int(input())
seen = set()
res = []
n = len(lst)
for i in range(n):
    for j in range(i+1,n):
        if lst[i] + lst[j] == k:
            if i not in seen:
                res.append((lst[i],lst[j]))
                seen.add((lst[i],lst[j]))

res.sort(key=lambda x: x[0])
print(res)
```

```python
#W6A12:
lst = list(map(int,input().split()))
count = {}
for i in lst:
    if i in count:
        count[i] += 1
    else:
        count[i] = 1
max_count = max(count.values())
ans = min(k for k,v in count.items() if v == max_count)
print(ans)
```

```python
#W6A19:
def max_index(lst):
    max_val = lst[0]
    idx = 0
    for i in range(1, len(lst)):
        if lst[i] > max_val:
            max_val = lst[i]
            idx = i
    return idx


lst = list(map(int, input().split()))
print(max_index(lst))
```

```python
#W6A20:
def find_k(lst, k):
    for i in range(len(lst)):
        if lst[i] == k:
            return i
    return -1


lst = list(map(int, input().split()))
k = int(input())
print(find_k(lst, k))
```
