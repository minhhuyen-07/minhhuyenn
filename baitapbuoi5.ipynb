```python
#1:
def max_of_two():
    a,b = map(int,input().split())
    return max(a,b)
print(max_of_two())
```

```python
#2:
def swap():
    a,b = map(int,input().split())
    return b,a
print(swap())

    
```

```python
#3:
from math import*
def nguyen_to():
    n = int(input())
    if n < 2:
        return False
    for i in range(2,int(sqrt(n)+1)):
        if n % i == 0:
            return False
    return True
print(nguyen_to())
```

```python
#4:
def so_hoan_hao():
    n = int(input())
    tong = 0
    for i in range(1,n):
        if n % i == 0:
            tong += i
    if tong == n:
        return True
    return False
print(so_hoan_hao())    
```

```python
#5:
def so_nguyen():
    lst = list(map(int,input().split()))
    k = int(input())
    for i in range(len(lst)):
        if lst[i] == k:
            return i + 1
    return -1
print(so_nguyen())
    
```

```python
#6:
def giai_thua():
    n = int(input())
    gt = 1
    for i in range(1,n+1):
        gt *= i
    return gt
print(giai_thua())
```

```python
#8:
def Hamming(x, y):
    x = x ^ y
    ans = 0
    while x > 0:
        ans += x % 2
        x = int(x / 2)
    return ans
```

```python
#9:
def tong(n):
    s = str(n)
    ans = 0
    for i in s:
        ans += int(i)
    return ans
```

```python
#10:
def tong_chu_so():
    n = int(input())
    tong = 0 
    while n > 0:
        tong += n % 10
        n //= 10
    return tong
print(tong_chu_so())

```
