```python
#W7A1:
def binary_search(arr, left, right, target):
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == target:
            return mid
        elif arr[mid] < target:
            left = mid + 1
        else:
            right = mid - 1
    return -1
```

```python
#W7A2:
def count_occurrences(arr, x):
    count = 0
    for i in arr:
        if i == x:
            count += 1
    return count
```

```python
#W7A3:
def bubble_sort(arr):
    a = arr[:]
    n = len(a)
    for i in range(n):
        for j in range(0, n - i - 1):
            if a[j] > a[j + 1]:
                a[j], a[j + 1] = a[j + 1], a[j]
    return a
```

```python
#W7A3:
def selection_sort(arr):
    a = arr[:]
    n = len(a)
    for i in range(n):
        min_idx = i
        for j in range(i + 1, n):
            if a[j] < a[min_idx]:
                min_idx = j
        a[i], a[min_idx] = a[min_idx], a[i]
    return a
```

```python
#W7A4:
def most_frequent(arr):
    freq = {}
    for x in arr:
        freq[x] = freq.get(x, 0) + 1

    max_count = 0
    result = None
    for x in arr:
        if freq[x] > max_count:
            max_count = freq[x]
            result = x
    return result, max_count
```

```python
#W7A5:
def count_pairs(arr, X):
    count = 0
    n = len(arr)
    for i in range(n):
        for j in range(i + 1, n):
            if arr[i] + arr[j] == X:
                count += 1
    return count
```

```python
#W7A6:
def flatten_list(lst):
    res = []
    for sub in lst:
        for x in sub:
            res.append(x)
    return res
```

```python
#W7A7:
def length_of_LIS(nums):
    if not nums:
        return 0
    dp = [1] * len(nums)
    for i in range(len(nums)):
        for j in range(i):
            if nums[j] < nums[i]:
                dp[i] = max(dp[i], dp[j] + 1)
    return max(dp)
```

```python
#W7A8:
def min_interval(intervals, queries):
    res = []
    for q in queries:
        ans = float('inf')
        for l, r in intervals:
            if l <= q <= r:
                ans = min(ans, r - l + 1)
        res.append(ans if ans != float('inf') else -1)
    return res
```

```python
#W7A9:
def count_smaller(nums):
    res = []
    for i in range(len(nums)):
        cnt = 0
        for j in range(i + 1, len(nums)):
            if nums[j] < nums[i]:
                cnt += 1
        res.append(cnt)
    return res
```

```python
#W7A10:
def max_subarray_modulo(nums, m):
    prefix = 0
    max_mod = 0
    prefix_set = [0]

    for x in nums:
        prefix = (prefix + x) % m
        max_mod = max(max_mod, prefix)
        for p in prefix_set:
            if p > prefix:
                max_mod = max(max_mod, (prefix - p + m) % m)
        prefix_set.append(prefix)
    return max_mod
```

```python
#W7A11:
def count_word(sentence, word):
    words = sentence.split()
    count = 0
    for w in words:
        if w == word:
            count += 1
    return count
```

```python
#W7A12:
def sort_dictionary_order(words):
    indexed = list(enumerate(words))
    indexed.sort(key=lambda x: x[1])
    return " ".join(chr(65 + i) for i, _ in indexed)
```

```python
#W7A13:
def longest_harmonious(arr):
    freq = {}
    for x in arr:
        freq[x] = freq.get(x, 0) + 1

    res = 0
    for x in freq:
        if x + 1 in freq:
            res = max(res, freq[x] + freq[x + 1])
    return res
```

```python
#W7A14:
def find_seven(arr):
    positions = []
    for i in range(len(arr)):
        if arr[i] == 7:
            positions.append(i)
    if not positions:
        return "Not found"
    return " ".join(map(str, positions[::-1]))
```
