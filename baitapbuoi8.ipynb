```python
#W8A1:
from math import*
class Cylinder:
    def __init__(self,r,h):
        self.r = r
        self.h = h
    def dien_tich_be_mat(self):
        return 2 * (self.r**2) * pi + 2 * self.h * self.r * pi
    def the_tich(self,):
        return 2 * pi * (self.r**2) * self.h
r,h = map(float,input().split())
c = Cylinder(r,h)
print(f"{c.dien_tich_be_mat():.2f}")
print(f"{c.the_tich():.2f}")
```

```python
#W8A2:
class Date:
    def __init__(self,d,m,y):
        self.d = d
        self.m = m
        self.y = y
    def nam_nhuan(self):
        return (self.y % 400 == 0) or (self.y % 4 == 0 and self.y % 100 != 0)
    def ngay_thang(self):
        if self.m in [1,3,5,7,8,10,12]:
            return 31
        if self.m in [4,6,9,11]:
            return 30
        if self.m == 2:
            if self.nam_nhuan():
                return 29
            return 28
    def is_valid(self):
        if self.y < 0:
            return False
        if self.m < 1 or self.m > 12:
            return False
        if self.d < 1 or self.d > self.ngay_thang():
            return False
        return True
    def ngay_tiep_theo(self):
        if not self.is_valid():
            return 'INVALID'
        if self.d < self.ngay_thang():
            self.d += 1
        else:
            self.d = 1
            if self.m < 12:
                self.m += 1
            else:
                self.m = 1
                self.y += 1
        return f"{self.d:02d}/{self.m:02d}/{self.y}"
s = input().strip()
d, m, y = map(int, s.split())

date = Date(d, m, y)

if not date.is_valid():
    print("INVALID")
else:
    print(date.ngay_tiep_theo())
```

```python
#W8A5:
class ShoppingCart:
    def __init__(self):
        self.items = []

    def add(self, name, price):
        self.items.append({"name": name, "price": price})

    def remove(self, name):
        self.items = [x for x in self.items if x["name"] != name]

    def is_empty(self):
        return len(self.items) == 0

    def total(self):
        return sum(x["price"] for x in self.items)

    def show(self):
        for x in self.items:
            print(x["name"], x["price"])

    def clear(self):
        self.items.clear()
```

```python
#W8A6:
class Rectangle:
    def __init__(self, w, h):
        self.w = w
        self.h = h

    def area(self):
        return self.w * self.h

    def perimeter(self):
        return 2 * (self.w + self.h)

    def scale(self, k):
        self.w *= k
        self.h *= k


w, h, k = map(float, input().split())
r = Rectangle(w, h)
r.scale(k)
print(f"{r.area():.2f} {r.perimeter():.2f}")
```

```python
#W8A7:
import math

class Fraction:
    def __init__(self, num, den):
        if den < 0:
            num, den = -num, -den
        g = math.gcd(num, den)
        self.num = num // g
        self.den = den // g

    def add(self, other):
        return Fraction(self.num*other.den + other.num*self.den,
                        self.den*other.den)

    def sub(self, other):
        return Fraction(self.num*other.den - other.num*self.den,
                        self.den*other.den)

    def mul(self, other):
        return Fraction(self.num*other.num, self.den*other.den)

    def div(self, other):
        return Fraction(self.num*other.den, self.den*other.num)

    def __str__(self):
        return f"{self.num}/{self.den}"


a, b, op, c, d = input().split()
f1 = Fraction(int(a), int(b))
f2 = Fraction(int(c), int(d))

if op == "+":
    print(f1.add(f2))
elif op == "-":
    print(f1.sub(f2))
elif op == "*":
    print(f1.mul(f2))
elif op == "/":
    print(f1.div(f2))
```

```python
#W8A8:
class BankAccount:
    def __init__(self, owner, balance):
        self.owner = owner
        self.balance = balance

    def deposit(self, x):
        self.balance += x

    def withdraw(self, x):
        if x <= self.balance:
            self.balance -= x


owner, bal = input().split()
acc = BankAccount(owner, float(bal))

n = int(input())
for _ in range(n):
    cmd, x = input().split()
    x = float(x)
    if cmd == "DEPOSIT":
        acc.deposit(x)
    else:
        acc.withdraw(x)

print(f"{acc.balance:.2f}")

```

```python
#W8A9:
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, subject, score):
        self.scores.append(score)

    def avg(self):
        return sum(self.scores) / len(self.scores)

    def rank(self):
        a = self.avg()
        if a >= 8:
            return "Excellent"
        if a >= 6.5:
            return "Good"
        if a >= 5:
            return "Average"
        return "Poor"


name = input().strip()
st = Student(name)
n = int(input())
for _ in range(n):
    _, s = input().split()
    st.add_score(_, float(s))

print(f"{name} {st.avg():.2f} {st.rank()}")
```

```python
#W8A10:
class Book:
    def __init__(self, title, author, year):
        self.title = title
        self.author = author
        self.year = year


class Library:
    def __init__(self):
        self.books = []

    def add(self, book):
        self.books.append(book)

    def count_by_author(self, name):
        return sum(1 for b in self.books if b.author == name)

    def find_by_year(self, y):
        return sum(1 for b in self.books if b.year == y)


lib = Library()
n = int(input())
for _ in range(n):
    s = input()
    if s.startswith("ADD"):
        _, rest = s.split(" ", 1)
        t, a, y = rest.split(";")
        lib.add(Book(t, a, int(y)))
    elif s.startswith("COUNTYEAR"):
        y = int(s.split()[1])
        print(lib.find_by_year(y))
    else:
        a = s.split()[1]
        print(lib.count_by_author(a))
```

```python
#W8A11:
import math

def vector(a, b):
    return [b[i] - a[i] for i in range(3)]

def cross(u, v):
    return [
        u[1]*v[2] - u[2]*v[1],
        u[2]*v[0] - u[0]*v[2],
        u[0]*v[1] - u[1]*v[0]
    ]

def dot(u, v):
    return sum(u[i]*v[i] for i in range(3))

def norm(u):
    return math.sqrt(dot(u, u))


A = list(map(float, input().split()))
B = list(map(float, input().split()))
C = list(map(float, input().split()))
D = list(map(float, input().split()))

n1 = cross(vector(A, B), vector(A, C))
n2 = cross(vector(B, C), vector(B, D))

angle = math.degrees(math.acos(dot(n1, n2) / (norm(n1) * norm(n2))))
print(f"{angle:.2f}")
```

```python
#W8A12:
class ComplexNumber:
    def __init__(self, r, i):
        self.r = r
        self.i = i

    def multiply(self, other):
        real = self.r*other.r - self.i*other.i
        imag = self.r*other.i + self.i*other.r
        return ComplexNumber(real, imag)


a, b = map(float, input().split())
c, d = map(float, input().split())
z1 = ComplexNumber(a, b)
z2 = ComplexNumber(c, d)
res = z1.multiply(z2)
print(f"{res.r:.0f} {res.i:.0f}")
```

```python
#W8A13:
class Triangle:
    def __init__(self, a, b, c):
        self.a, self.b, self.c = a, b, c

    def valid(self):
        return (self.a + self.b > self.c and
                self.a + self.c > self.b and
                self.b + self.c > self.a)

    def getArea(self):
        if not self.valid():
            return -1
        p = (self.a + self.b + self.c) / 2
        return math.sqrt(p*(p-self.a)*(p-self.b)*(p-self.c))


a, b, c = map(int, input().split())
t = Triangle(a, b, c)
area = t.getArea()
if area == -1:
    print("invalid")
else:
    print(f"{area:.2f}")
```
