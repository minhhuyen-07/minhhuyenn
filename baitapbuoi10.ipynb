```python
#W10A3:
def happy_numbers(n):
    kq = []
    while n != 1 and n not in kq:
        kq.append(n)
        n = sum(int(c) ** 2 for c in str(n))
    return n == 1
try:
    a = int(input())
    if happy_numbers(a):
        print("YES")
    else:
        print("NO")
except ValueError:
    print("NO")
    
```

```python
#W10A11:
N = int(input())
a = list(map(int, input().split()))
if N <= 0 or len(a) != N:
    print("Mang khong hop le")
else:
    hop_le = True
    for i in range(N):
        if a[i] <= 0 or a.count(a[i]) > 1:
            hop_le = False
            break

    if hop_le:
        print(sum(a))
    else:
        print("Mang khong hop le")
```
