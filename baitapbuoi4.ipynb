```python
#8:
def check(n):
    s = str(n)
    s1 = s[::-1]
    return s == s1

def reve(n):
    s = str(n)
    s1 = s[::-1]
    return int(s1)

n = int(input())
cnt = 0
while check(n) == False:
    cnt += 1
    n = n + reve(n)
print(cnt, n)
```

```python
#9:
import math

def check(n):
    li = {"&"}
    s = str(n)
    for i in s:
        li.add(i)
    return len(li) - 1 == len(s)

n = int(input())
i = 1
while i <= int(math.sqrt(n)):
    if check(i * i):
        print(i * i, end = ' ')
    i += 1
```

```python
#10:
mp = {1: 1}

def ql(n):
    if n in mp: 
        return mp[n]
    if n % 2 == 0:
        mp[n] = ql(n / 2) + 1
    else:
        mp[n] = ql(n * 3 + 1) + 1
    return mp[n]

n = int(input())
ans = 0
mx = 0
for i in range(1, n + 1):
    if ql(i) > mx:
        ans = i
        mx = ql(i)
print(ans, mx)
```
